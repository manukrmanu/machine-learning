```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.30,random_state=42)
x_test
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_test
```

```python
x_train
```

```python
#modeling
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from types import ClassMethodDescriptorType
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
label=['iris-setosa','iris-verginica','iris-virginica']
cmd=confusion_matrix(y_test,y_pred)
dis=ConfusionMatrixDisplay(cmd,display_labels=label)
dis.plot()
score=accuracy_score(y_test,y_pred)
print(cmd)
score
```
