```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/heart.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x.ndim
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
clf=KNeighborsClassifier()
param={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
gd=GridSearchCV(clf,param,cv=10,scoring='accuracy')
gd.fit(x_train,y_train)
print(gd.best_params_)
```

```python
modelnew=KNeighborsClassifier(n_neighbors=9,weights='distance')
modelnew.fit(x_train,y_train)
y_prednew=modelnew.predict(x_test)
print(classification_report(y_test,y_prednew))
```
