```python
import matplotlib.pyplot as plt
x=[150,155,156,157]
y=[45,50,51,58]
plt.plot(x,y,'*:r')
plt.grid(color='green',linestyle=':',linewidth=1)
```

```python
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1)
plt.plot(x1,y1,)
plt.title('graph1')
plt.xlabel('xgraph1')
plt.ylabel('ygraph1')

x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('graph2')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.suptitle('matplotlib')

```
