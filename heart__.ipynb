```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/heart_missing2.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
x=df['cp'].mode()[0]
df['cp'].fillna(x,inplace=True)
y=df['fbs'].mode()[0]
df['fbs'].fillna(y,inplace=True)
z=df['oldpeak'].mode()[0]
df['oldpeak'].fillna(z,inplace=True)
xx=df['restecg'].max()
df['restecg'].fillna(xx,inplace=True)
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_test
```

```python
#modeling
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
k_model=KNeighborsClassifier(n_neighbors=7)
n_model=GaussianNB()
s_model=SVC()
list_model=[k_model,n_model,s_model]
```

```python
for i in list_model:
  print("model==",i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  y_pred
  print(confusion_matrix(y_test,y_pred))
  print(accuracy_score(y_test,y_pred))
  print("*"*100)

```
