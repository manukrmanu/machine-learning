```python
import numpy as np
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[8,9,10,11]
plt.plot(x,y)
plt.show()
```

```python
x=[5,2,3,4]
y=[8,9,10,11]
plt.plot(x,y,"*")
```

```python
x=np.array([2,4,6,8])
y=np.array([6,8,5,9])
plt.plot(x,y,"*")
```

```python
data=np.array([3,8,1,10,5,7])
plt.plot(data,"*")
```

```python
y=np.array([3,8,1,10,7,6,2])
plt.plot(y,marker="o")
```

```python
plt.plot(y,'*--g')
```

```python
plt.plot(y,marker='o',ms=15)
```

```python
plt.plot(y,marker='o',mfc='r',mec='y')
```

```python
plt.plot(y,linestyle='dotted',linewidth=5)
```

```python
x1=[1,2,6,4]
y1=[4,7,9,2]
x2=[3,4,6,1]
y2=[1,7,2,8]
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
x=[150,155,156,157]
y=[45,50,51,58]
plt.plot(x,y)
dic1={'family':'serif','color':'b','size':'20'}
dic2={'family':'serif','color':'b','size':'15'}
plt.xlabel('height in cm',fontdict=dic2)
plt.ylabel('weight in kg',fontdict=dic2)
plt.title('height & weight',fontdict=dic1)
```

```python
x=[20,32,39,49,57]
y=[50,45,20,30,35]
plt.plot(x,y)
plt.xlabel('height in cm',fontdict=dic2)
plt.ylabel('weight in kg',fontdict=dic2)
plt.title('height & weight',fontdict=dic1)

```
